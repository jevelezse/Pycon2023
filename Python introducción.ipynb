```python
def gc(dna):
    "Esta función computa el procentaje GC"
    nbases=dna.count('n')+dna.count('N')
    gcporcentaje=float(dna.count('c')+dna.count('C')+dna.count('g')+dna.count('G'))*100.0/(len(dna)-nbases)
    return gcporcentaje   
    
```

```python
dna='ATCGNNNCCCC'
gc(dna)
```

```python
def count1(dna,base):
    i=0
    for c in dna:
        if c == base:
            i += 1
    return i
```

```python
base='c'
count1(dna,base)
```

```python
def count2(dna,base):
    i=0
    for j in range(len(dna)):
        if dna[j] == base:
            i += 1
    return i
```

```python
def count3(dna,base):
    match = [c == base for c in dna]
    return sum(match)
    
        
```

```python
f=open('fasta.txt')
seqs={}
for line in f:
    line=line.rstrip()
    if line.startswith('>'):
        words=line.split()
        name=words[0][1:]
        seqs[name]=''
    else:
        seqs[name] = seqs[name]+line
f.close()
print(seqs)
```
